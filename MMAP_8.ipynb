# **Описание задачи.**

Решить интегральное уравнение Вольтерра 2-го рода с вырожденным ядром с помощью методов трапеций, Симпсона, трёх восьмых для числа разбиений $n=10$, $N=100$

$$u(x) - \int \limits_{0}^{x} e^{x - s} u(s) ds = e^x$$



# **Численная схема описанного математического метода.**

## Метод трапеций:



$$I_{трапеции} = \frac{h}{2} \sum^{n}_{i=0} k_i f(x_i); h = \frac{b - a}{n}, x_i = a + ih, i=0,_\cdots,n$$

\begin{equation*}
    k_i = \begin{cases}
        1, i = 0
        \\
        2, i \neq 0, i \neq n-1
        \\
        1, i = n
    \end{cases}
\end{equation*}

$$R_{трапеции} = -\frac{(b - a)^3}{12n^3} f''(\xi), a<\xi<b$$

## Метод Симпсона:



$$I_{Симпсон} = \frac{h}{3} \sum^{n}_{i=0} k_i f(x_i); h = \frac{b - a}{n}, x_i = a + ih, i=0,_\cdots,n$$

\begin{equation*}
    k_i = \begin{cases}
        1, i = 0
        \\
        2, i\ \vdots\ 2
        \\
        4, i \not{\vdots}\ 2
        \\
        1, i = n
    \end{cases}
\end{equation*}

$$R_{Симпсон} = -\frac{(b - a)^5}{2880n^4} f^{(4)}(\xi), a<\xi<b$$

## Метод трёх восьмых:



$$I_{\frac{3}{8}} = \frac{3h}{8} \sum^{n}_{i=0} k_i f(x_i); h = \frac{b - a}{n}, x_i = a + ih, i=0,_\cdots,n$$

\begin{equation*}
    k_i = \begin{cases}
        1, i = 0
        \\
        2, i\ \vdots\ 3
        \\
        3, i \not{\vdots}\ 3
        \\
        1, i = n
    \end{cases}
\end{equation*}

$$R_{\frac{3}{8}} = -\frac{(b - a)^5}{80n^4} f^{(4)}(\xi), a<\xi<b$$

# **Аналитическое решение.**

$$u(x) - \int \limits_{0}^{x} e^{x - s} u(s) ds = e^x$$

$$u(x) - \int \limits_{0}^{x} e^x \cdot e^{-s} u(s) ds = e^x$$

$$u(x) - e^x \int \limits_{0}^{x} e^{-s} u(s) ds = e^x$$

$$u(x) = e^x + e^x \int \limits_{0}^{x} e^{-s} u(s) ds = e^x (1 + \int \limits_{0}^{x} e^{-s} u(s) ds)$$

$$\sqsupset y(x) = \frac{u(x)}{e^x} \Rightarrow y = 1 + \int \limits_{0}^{x} e^{-s} u(s) ds, y(0)=1$$

$$y' = e^{-x} u(x) = e^{-x} \cdot e^x \cdot y = y$$

$$\frac{dy}{dx} = y$$

$$\frac{dy}{y} = dx$$

$$\int \frac{dy}{y} = \int dx$$

$$ln(y) = x + C$$

$$y = exp(x + C) = e^x \cdot e^C = e^x \cdot \tilde{C}, \tilde{C} = e^C$$

$$1 = e^0 \cdot \tilde{C} \Rightarrow \tilde{C} = 1$$

$$y = e^x$$

$$y(x) = \frac{u(x)}{e^x} \Rightarrow u(x) = e^{2x}$$



# **Программная реализация.**

Что нужно сделать: сравнить аналитическое решение $u_1(x) = e^{2x}$ с численным $u_2(x) = e^x (1 + \int \limits_{0}^{x} e^{-s} u(s) ds)$, где интеграл $\int \limits_{0}^{x} e^s ds$ мы считаем с помощью ЧМов. Верхнюю границу интегрирования лучше ставить как можно мешьше, чтоб ошибка была тоже очень маленькой (я взял $x=0.1$)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")

import warnings
warnings.filterwarnings("ignore")
```

```python
# Иксы
def x(i):
    return a + i * h

# Подынтегральная функция
def f(x):
    return np.exp(x)

# Аналитическое решение
def u_analytical(x):
    return np.exp(2 * x)

# Численное решение
def u_numerical(x, C_Int):
    return np.exp(x) * (1 + C_Int)

# Ошибка
def delta(u_ana, u_num):
    return (np.sum((u_ana - u_num) ** 2)) / (np.sum(u_ana ** 2))

# Графика
def visualization(X, u_ana, u_num, delta, method_name):
    plt.plot(X, u_ana, c='green', label='Аналитическое решение')
    plt.plot(X, u_num, c='red', label=method_name, linestyle='--')
    plt.title(f'Визуализация решений; $\delta$={delta}')
    plt.legend()
```

```python
# Коэффициент к
def k_trapezoid(i):
    return 1 * (i == 0 or i == n) + 2 * (i != 0 and i != n)

# Метод трапеций
def Int_trapezoid(n, h):
    S = f(x(0)) + f(x(n))
    for i in range(1, n):
        S += k_trapezoid(i) * f(x(i))
    S = (h / 2) * S

    return S
```

```python
# Коэффициент к
def k_Simpson(i):
    return 1 * (i == 0 or i == n) + 2 * (i % 2 == 0 and (i != 0 or i != n)) + 4 * (i % 2 != 0 and (i != 0 or i != n))

# Метод Симпсона
def Int_Simpson(n, h):
    S = f(x(0)) + f(x(n))
    for i in range(1, n):
        S += k_Simpson(i) * f(x(i))
    S = (h / 3) * S

    return S
```

```python
# Коэффициент к
def k_three_eighths(i):
    return 1 * (i == 0 or i == n) + 2 * (i % 3 == 0 and (i != 0 or i != n)) + 3 * (i % 3 != 0 and (i != 0 or i != n))

# Метод трёх восьмых
def Int_three_eighths(n, h):
    S = f(x(0)) + f(x(n))
    for i in range(1, n):
        S += k_three_eighths(i) * f(x(i))
    S = (3 / 8) * h * S

    return S
```

```python
a, b = 0, 0.1
```

# **Оценка погрешности расчётов для каждого метода и каждого шага.**

## n = 10

```python
n = 10
h = (b - a) / n
```

```python
C_trapezoid = Int_trapezoid(n, h)
C_Simpson = Int_Simpson(n, h)
C_three_eighths = Int_three_eighths(n, h)
```

```python
X = np.array([x(i) for i in range(n + 1)])

U_analitycal = u_analytical(X)
U_trapezoid = u_numerical(X, C_trapezoid)
U_Simpson = u_numerical(X, C_Simpson)
U_three_eighths = u_numerical(X, C_three_eighths)
```

### Метод трапеций

```python
visualization(X, U_analitycal, U_trapezoid, delta(U_analitycal, U_trapezoid), 'Метод трапеций')
```

### Метод Симпсона

```python
visualization(X, U_analitycal, U_Simpson, delta(U_analitycal, U_Simpson), 'Метод Симпсона')
```

### Метод трёх восьмых

```python
visualization(X, U_analitycal, U_three_eighths, delta(U_analitycal, U_three_eighths), 'Метод трёх восьмых')
```
